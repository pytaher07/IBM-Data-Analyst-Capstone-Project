## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt

df['ConvertedComp'].plot(kind='box', figsize=(8,5))
plt.title('Distribution Data of Annual Compensation')
plt.ylabel('Annual Compensation')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
df['ConvertedComp'].describe()
```

Find out the upper and lower bounds.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

print('The Upper bound of the Data: ', upper_bound)
print('The Lower Bound of the Data: ', lower_bound)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
Outliers = []
for i in df['ConvertedComp']:
    if ((i > upper_bound) or (i < lower_bound)):
        Outliers.append(i)
print('The Number of Outliers in the Data: ', len(Outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df_No_Outliers = df['ConvertedComp'][~((df['ConvertedComp'] < lower_bound)|
                                       (df['ConvertedComp'] > upper_bound))]

df_No_Outliers.describe()
```

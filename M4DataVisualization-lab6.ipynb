# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')

QUERY = """
SELECT Respondent, ConvertedComp
FROM master
"""

df_ConvertedComp = pd.read_sql_query(QUERY, conn)

df_ConvertedComp['ConvertedComp'].plot(kind='hist', figsize=(8,5))

plt.title('Histogram of Converted COmpensation')
plt.ylabel('Number of Respondents')
plt.xlabel('ANNUAL Salaries')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Respondent, Age 
FROM master
"""
df_Age = pd.read_sql_query(QUERY, conn)
print(df_Age['Age'].describe())

df_Age['Age'].plot(kind='box', figsize=(8,5))

plt.title("Box Plot of Respondents' Age")
plt.ylabel('Age Intervals')

plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df_age_workweekhrs = pd.read_sql_query(QUERY, conn)
df_age_workweekhrs.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(8,5), color='blue')

plt.title('Relation Between Age and Working Hours per Week of Respondents')
plt.xlabel('Age')
plt.ylabel('Week Working Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df_codeReview = pd.read_sql_query(QUERY, conn)

df_code_review.info()

#normalize the age column for better visual of the size bubble

norm_age = (df_codeReview['Age'] - df_codeReview['Age'].min()) / (df_codeReview['Age'].max() - df_codeReview['Age'].min())

df_codeReview.plot(kind='scatter',
               x='WorkWeekHrs',
               y='CodeRevHrs',
               figsize=(10, 6),
               alpha=0.5,  # transparency
               color='red',
               s = norm_age *1000 +10,  # pass in weights 
               )

plt.title('Relation Between WorkWeekHrs and CodeRevHrs of Respondents')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc='upper right', fontsize='x-large')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT * FROM DatabaseDesireNextYear
"""

df_dbDesire = pd.read_sql_query(QUERY, conn)

df_dbDesire.head()
```

```python
df_groupby= df_dbDesire.groupby('DatabaseDesireNextYear', axis=0).count()

df_groupby.sort_values(['Respondent'], ascending=False, inplace=True)

df_groupby.head()
```

```python
df_Piechart= df_groupby.head()

df_Piechart.plot(kind='pie',
                figsize=(12, 6),
                autopct='%1.1f%%', 
                startangle=90,    
                shadow=True,       
                labels=None,         # turn off labels on pie chart
                pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                subplots=True
                )

plt.title('Top 5 Databases Respondents Wish To Learn Next Year', y=1.12)
plt.axis('equal')
plt.legend(labels= df_Piechart.index, loc='upper left')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stacked = pd.read_sql_query(QUERY,conn)

df_stacked.info()
```

```python
df_stacked = df_stacked.loc[(df_stacked['Age'] >= 30) & (df_stacked['Age'] <= 35)].reset_index().drop(columns=['index'])
df_stacked
```

```python
import matplotlib.pyplot as plt

df_stacked.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind='bar', stacked=True, figsize=(25, 10))


plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35',
          loc='center',pad=30, fontsize=30, color='blue')
plt.xlabel('CodeRevHrs', fontsize = 20)
plt.ylabel('WorkWeekHrs', fontsize = 20)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, ncol=1, title='Age')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT ConvertedComp, Age
FROM master
"""
df_line = pd.read_sql_query(QUERY,conn)

df_line.info()
```

```python
df_line = df_line.loc[(df_line['Age'] >= 45) & (df_line['Age'] <= 60)].reset_index().drop(columns=['index'])
df_line.dropna(inplace=True)
df_line
```

```python
df_line = df_line.groupby('Age', axis=0).median()
df_line
```

```python
df_line.plot(kind='line')

plt.title('Average Compensation Based on Age Groups')
plt.ylabel('Converted Compensation (Annual USD')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`
